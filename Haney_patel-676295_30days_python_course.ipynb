# Day 3 - Operators

Operators are symbols used to perform operations on variables and values. Python supports arithmetic, comparison, logical, assignment, identity, and membership operators.

````python
# Arithmetic operators
# Used to perform mathematical operations

a, b = 15, 4

print('Addition:', a + b)
print('Subtraction:', a - b)
print('Multiplication:', a * b)
print('Division:', a / b)          # always returns float
print('Floor division:', a // b)   # drops the remainder
print('Modulus:', a % b)           # remainder only
print('Exponentiation:', a ** b)   # 15 to the power of 4
````

````python
# Comparison operators
# Return True or False

x = 25

print('Equal (==):', x == 25)
print('Not equal (!=):', x != 24)
print('Greater than (>):', x > 24)
print('Less than (<):', x < 30)
print('Greater or equal (>=):', x >= 25)
print('Less or equal (<=):', x <= 25)
````

````python
# Logical operators
# and, or, not - used to combine boolean values

a = True
b = False

print('True and False:', a and b)
print('True or False:', a or b)
print('not True:', not a)

# Practical example
age = 25
has_license = True
can_drive = age >= 18 and has_license
print(f'\nCan drive: {can_drive}')
````

````python
# Bitwise operators
# Work on binary representations of numbers

x, y = 6, 3  # 110 and 011 in binary

print('Bitwise AND:', x & y)    # 010 = 2
print('Bitwise OR:', x | y)     # 111 = 7
print('Bitwise XOR:', x ^ y)    # 101 = 5
print('Left shift:', x << 1)    # 1100 = 12
print('Right shift:', x >> 1)   # 011 = 3
````

````python
# Assignment and other operators

n = 10
print('Start:', n)

n += 5   # n = n + 5
print('After += 5:', n)

n -= 3
print('After -= 3:', n)

n *= 2
print('After *= 2:', n)

n //= 4
print('After //= 4:', n)

# Identity operators - check if objects are the same
my_list = [1, 2, 3]
x = my_list
y = [1, 2, 3]

print('\nx is my_list:', x is my_list)    # True - same object
print('y is my_list:', y is my_list)    # False - different object

# Membership operators
print('3 in my_list:', 3 in my_list)
print('5 not in my_list:', 5 not in my_list)
````

# Day 5 - Lists

````python
# Creating lists

empty_list = []
numbers = [1, 2, 3, 4, 5]
fruits = ['mango', 'banana', 'papaya', 'orange']
mixed = ['Haney', 19, True, 3.14, ['nested', 'list']]

print('Numbers:', numbers)
print('Fruits:', fruits)
print('Mixed:', mixed)
print('Length of fruits:', len(fruits))
````

````python
# List methods
# Methods that modify or query lists

courses = ['Python', 'JavaScript', 'Java']

courses.append('C++')
print('After append:', courses)

courses.insert(1, 'HTML')
print('After insert:', courses)

courses.remove('HTML')
print('After remove:', courses)

popped = courses.pop()
print('Popped item:', popped)
print('After pop:', courses)

courses.sort()
print('After sort:', courses)

courses.reverse()
print('After reverse:', courses)

print('Index of Python:', courses.index('Python'))
print('Count of Python:', courses.count('Python'))
````

````python
# List operations and unpacking

a = [1, 2, 3]
b = [4, 5, 6]

print('Concatenation:', a + b)
print('Repetition:', a * 3)
print('5 in b:', 5 in b)

# Copy a list (important distinction)
original = [1, 2, 3, 4, 5]
copy = original.copy()  # or list(original) or original[:]
copy.append(99)

print('Original:', original)
print('Copy:', copy)

# Unpacking
first, *middle, last = [10, 20, 30, 40, 50]
print(f'First: {first}, Middle: {middle}, Last: {last}')
````

````python
# 2D lists (matrices)

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print('Matrix:')
for row in matrix:
    print(row)

print('\nElement at [1][2]:', matrix[1][2])

# Flatten a 2D list
flat = [item for row in matrix for item in row]
print('Flattened:', flat)
````

# Day 7 - Sets

Sets are unordered collections of unique items. Sets are mutable and defined using curly braces or the set() constructor. Sets do not allow duplicate values.

````python
# Creating sets
# Duplicates are automatically removed

my_set = {1, 2, 3, 4, 5}
print('Set:', my_set)

fruits = set(['apple', 'banana', 'mango', 'apple'])  # duplicate removed
print('Fruits set:', fruits)

empty_set = set()  # use set() not {} for empty set
print('Empty set:', empty_set, type(empty_set))
````

````python
# Set methods

languages = {'Python', 'Java', 'C++'}

languages.add('JavaScript')
print('After add:', languages)

languages.remove('Java')
print('After remove:', languages)

languages.discard('Ruby')  # no error if not found
print('After discard:', languages)

print('Length:', len(languages))
````

````python
# Set operations - union, intersection, difference
# Like mathematical set operations

A = {1, 2, 3, 4, 5}
B = {3, 4, 5, 6, 7}

print('Union (A | B):', A | B)
print('Intersection (A & B):', A & B)
print('Difference (A - B):', A - B)
print('Symmetric difference (A ^ B):', A ^ B)

# Subset and superset
print('A is subset of B:', A.issubset(B))
print('A is superset of {1, 2}:', A.issuperset({1, 2}))

# Disjoint - no common elements
even = {2, 4, 6, 8}
odd = {1, 3, 5, 7}
print('Even and odd are disjoint:', even.isdisjoint(odd))
````

````python
# Practical set example

my_skills = {'Python', 'HTML', 'CSS', 'Git'}
new_skills = {'Python', 'React', 'Node', 'Django'}

# Skills I already know that are in new list
common = my_skills & new_skills
print('Common skills:', common)

# All skills combined
all_skills = my_skills | new_skills
print('All skills:', all_skills)

# Skills I have but not in new list
unique = my_skills - new_skills
print('Unique to me:', unique)

# Unique ages in a group (removes duplicates)
ages = {19, 20, 19, 21, 20, 22, 19}
print('Unique ages:', ages)
````

# Day 8 - Dictionaries

Dictionaries are unordered collections of key-value pairs. Dictionaries are mutable and defined using curly braces with key:value pairs. Keys must be unique and immutable.

````python
# Creating dictionaries

student = {
    'name': 'Haney Patel',
    'age': 19,
    'country': 'Kenya',
    'city': 'Nairobi',
    'is_enrolled': True,
    'courses': ['Python', 'Data Structures', 'Algorithms']
}

print('Dictionary:', student)
print('Name:', student['name'])
print('Age:', student['age'])
print('Courses:', student['courses'])
````

````python
# Dictionary methods

person = {
    'name': 'Haney',
    'age': 19,
    'city': 'Nairobi'
}

# get() - safer than direct access
print('Name:', person.get('name'))
print('Phone:', person.get('phone', 'Not provided'))

# keys, values, items
print('Keys:', list(person.keys()))
print('Values:', list(person.values()))
print('Items:', list(person.items()))

# Adding and updating
person['email'] = 'haney@example.com'
person['age'] = 20
print('Updated:', person)
````

````python
# Removing items and checking keys

info = {'a': 1, 'b': 2, 'c': 3, 'd': 4}

info.pop('b')
print('After pop b:', info)

del info['c']
print('After del c:', info)

# Check if key exists
print('a in info:', 'a' in info)
print('z in info:', 'z' in info)
````

````python
# Nested dictionaries

class_data = {
    'student1': {'name': 'Haney', 'age': 19, 'grade': 'A'},
    'student2': {'name': 'John', 'age': 20, 'grade': 'B+'},
    'student3': {'name': 'Mary', 'age': 18, 'grade': 'A-'}
}

for key, student in class_data.items():
    print(f"{key}: {student['name']}, Age {student['age']}, Grade {student['grade']}")
````

# Day 19 - Classes and Objects

Object-Oriented Programming (OOP) uses classes and objects. A class is a blueprint for creating objects. An object is an instance of a class.

````python
# Basic class

class Student:
    school = 'Python University'  # class attribute

    def __init__(self, name, age, major):  # constructor
        self.name = name
        self.age = age
        self.major = major
        self.grades = {}

    def add_grade(self, course, grade):
        self.grades[course] = grade

    def get_gpa(self):
        if not self.grades:
            return 0.0
        grade_map = {'A+': 4.0, 'A': 4.0, 'B+': 3.3, 'B': 3.0, 'C': 2.0, 'F': 0.0}
        points = [grade_map.get(g, 0) for g in self.grades.values()]
        return sum(points) / len(points) if points else 0

    def __str__(self):
        return f'Student({self.name}, age={self.age}, major={self.major})'

# Creating objects
student1 = Student('Haney', 19, 'Computer Science')
student1.add_grade('Python', 'A')
student1.add_grade('Calculus', 'B+')

print(student1)
print('Grades:', student1.grades)
print(f'GPA: {student1.get_gpa():.2f}')
print('School:', Student.school)
````

````python
# Inheritance

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        return f'Hi, I am {self.name}, {self.age} years old.'


class Employee(Person):  # inherits from Person
    def __init__(self, name, age, employee_id, department):
        super().__init__(name, age)
        self.employee_id = employee_id
        self.department = department

    def introduce(self):  # overriding parent method
        base = super().introduce()
        return f'{base} I work in {self.department}. ID: {self.employee_id}.'


emp = Employee('Haney', 19, 'E001', 'Engineering')
print(emp.introduce())
````

````python
# Magic methods (dunder methods)

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f'Point({self.x}, {self.y})'

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Point(self.x - other.x, self.y - other.y)

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y


p1 = Point(3, 4)
p2 = Point(1, 2)
print('p1:', p1)
print('p2:', p2)
print('p1 + p2:', p1 + p2)
print('p1 - p2:', p1 - p2)
print('p1 == p2:', p1 == p2)
````

# Day 21 - Web Scraping

Web scraping involves extracting data from websites. Always check robots.txt and respect the website's terms of service.

````python
# Web scraping example with HTML parsing

from html.parser import HTMLParser

class SimpleHTMLParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.links = []
        self.headings = []

    def handle_starttag(self, tag, attrs):
        if tag == 'a':
            for attr in attrs:
                if attr[0] == 'href':
                    self.links.append(attr[1])
        if tag in ('h1', 'h2', 'h3'):
            self.current_tag = tag

    def handle_data(self, data):
        if hasattr(self, 'current_tag') and data.strip():
            self.headings.append(data.strip())
            del self.current_tag

sample_html = '''
<html>
  <h1>30 Days of Python</h1>
  <h2>Web Scraping</h2>
  <p>Learning web scraping</p>
  <a href="https://python.org">Python</a>
  <a href="https://github.com">GitHub</a>
</html>
'''

parser = SimpleHTMLParser()
parser.feed(sample_html)
print('Headings:', parser.headings)
print('Links:', parser.links)
````

````python
# Frequency distribution
from collections import Counter

scores = [72, 85, 90, 88, 76, 92, 68, 84, 79, 95, 88, 71]

def grade_bucket(score):
    if score >= 90: return 'A (90-100)'
    elif score >= 80: return 'B (80-89)'
    elif score >= 70: return 'C (70-79)'
    elif score >= 60: return 'D (60-69)'
    else: return 'F (<60)'

buckets = [grade_bucket(s) for s in scores]
freq = Counter(buckets)

print('Grade Distribution:')
for grade in ['A (90-100)', 'B (80-89)', 'C (70-79)', 'D (60-69)', 'F (<60)']:
    count = freq.get(grade, 0)
    bar = '#' * count
    print(f'{grade}: {bar} ({count})')
````

````python
# API Client class

class APIClient:
    def __init__(self, base_url):
        self.base_url = base_url
        self._mock_data = {
            'users': [
                {'id': 1, 'name': 'Haney', 'age': 19},
                {'id': 2, 'name': 'John', 'age': 20},
            ]
        }

    def get(self, endpoint):
        resource = endpoint.strip('/')
        return self._mock_data.get(resource, [])

    def post(self, endpoint, data):
        resource = endpoint.strip('/')
        if resource in self._mock_data:
            new_item = {'id': len(self._mock_data[resource]) + 1, **data}
            self._mock_data[resource].append(new_item)
            return new_item
        return None

# Using the API client
client = APIClient('https://api.example.com')

print('GET /users:', client.get('/users'))
new_user = client.post('/users', {'name': 'Mary', 'age': 24})
print('POST new user:', new_user)
print('After POST:', client.get('/users'))
````

# Day 27 - Decorators and Advanced Topics

Decorators are functions that modify other functions. Generators are memory efficient iterators.

````python
# Decorators - functions that modify other functions

def timing_decorator(func):
    import time
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f'{func.__name__} took {end-start:.6f} seconds')
        return result
    return wrapper

@timing_decorator
def calculate_sum(n):
    return sum(range(n))

result = calculate_sum(1000000)
print(f'Sum result: {result}')
````

````python
# Generator functions - memory efficient

def number_generator(n):
    for i in range(n):
        yield i

# Using generator
gen = number_generator(5)
for num in gen:
    print(num, end=' ')
print()

# Generator expression
squares_gen = (x**2 for x in range(5))
print('\nSquares from generator:', list(squares_gen))
````

# Day 29 - Working with External APIs

Practical example of consuming real-world APIs.

````python
# API Consumer Class

import json

class StudentAPI:
    def __init__(self):
        self.students = [
            {'id': 1, 'name': 'Haney', 'age': 19, 'major': 'CS', 'gpa': 3.85},
            {'id': 2, 'name': 'John', 'age': 20, 'major': 'Math', 'gpa': 3.9},
            {'id': 3, 'name': 'Mary', 'age': 24, 'major': 'Physics', 'gpa': 3.75}
        ]

    def get_all(self):
        return self.students

    def get_by_id(self, student_id):
        return next((s for s in self.students if s['id'] == student_id), None)

    def get_by_major(self, major):
        return [s for s in self.students if s['major'] == major]

    def get_high_gpa(self, threshold=3.8):
        return [s for s in self.students if s['gpa'] >= threshold]

    def add_student(self, name, age, major, gpa):
        new_id = max([s['id'] for s in self.students]) + 1
        new_student = {'id': new_id, 'name': name, 'age': age, 'major': major, 'gpa': gpa}
        self.students.append(new_student)
        return new_student

# Using the API
api = StudentAPI()

print('All students:', api.get_all())
print('\nStudent 1:', api.get_by_id(1))
print('\nCS students:', api.get_by_major('CS'))
print('\nHigh GPA students:', api.get_high_gpa())

new_student = api.add_student('Lisa', 23, 'Biology', 3.8)
print('\nAdded student:', new_student)
print('Total students:', len(api.get_all()))
````

````python
# Complete Student Management System

import json
from datetime import datetime

class StudentManagementSystem:
    def __init__(self, filename='/tmp/students.json'):
        self.filename = filename
        self.students = self.load_students()

    def load_students(self):
        try:
            with open(self.filename, 'r') as f:
                return json.load(f)
        except FileNotFoundError:
            return []

    def save_students(self):
        with open(self.filename, 'w') as f:
            json.dump(self.students, f, indent=4)

    def add_student(self, name, age, major):
        student_id = max([s['id'] for s in self.students], default=0) + 1
        student = {
            'id': student_id,
            'name': name,
            'age': age,
            'major': major,
            'grades': {},
            'date_added': datetime.now().isoformat()
        }
        self.students.append(student)
        self.save_students()
        return student

    def get_student(self, student_id):
        return next((s for s in self.students if s['id'] == student_id), None)

    def add_grade(self, student_id, course, grade):
        student = self.get_student(student_id)
        if student:
            student['grades'][course] = grade
            self.save_students()
            return True
        return False

    def get_students_by_major(self, major):
        return [s for s in self.students if s['major'] == major]

    def get_all_students(self):
        return self.students

    def get_student_count(self):
        return len(self.students)

# Demo
sms = StudentManagementSystem()

print('=== Student Management System ===')
print(f'Current students: {sms.get_student_count()}')

# Add students
s1 = sms.add_student('Haney', 19, 'Computer Science')
s2 = sms.add_student('John', 20, 'Mathematics')
s3 = sms.add_student('Mary', 24, 'Physics')

print(f'\nAdded 3 students. Total: {sms.get_student_count()}')

# Add grades
sms.add_grade(1, 'Python', 'A')
sms.add_grade(1, 'Data Structures', 'A-')
sms.add_grade(2, 'Calculus', 'B+')

print('\nCS Students:')
for student in sms.get_students_by_major('Computer Science'):
    print(f"  {student['name']}: {student['grades']}")

print(f'\nTotal students saved to {sms.filename}')
````
