## 📘 Day 3 — Operators

All the different operators in Python. 

````python
# Arithmetic operators
a, b = 15, 4

print('Addition:       ', a + b)
print('Subtraction:    ', a - b)
print('Multiplication: ', a * b)
print('Division:       ', a / b)    # always returns float
print('Floor division: ', a // b)   # drops remainder
print('Modulus:        ', a % b)    # remainder only
print('Exponentiation: ', a ** b)   # 15^4
````

````python
# Comparison operators -- return True or False
x = 19  # my age

print(x == 19)   # equal
print(x != 18)   # not equal
print(x > 18)    # greater than
print(x < 21)    # less than
print(x >= 19)   # greater than or equal
print(x <= 19)   # less than or equal
````

````python
# Logical operators
a = True
b = False

print('and:', a and b)
print('or: ', a or b)
print('not:', not a)

# Practical example
age = 19
has_id = True
can_enter_club = age >= 18 and has_id
print(f'Can enter club: {can_enter_club}')

# Bitwise operators
x, y = 6, 3   # 110 and 011 in binary
print('Bitwise AND:', x & y)   # 010 = 2
print('Bitwise OR: ', x | y)   # 111 = 7
print('Bitwise XOR:', x ^ y)   # 101 = 5
print('Left shift: ', x << 1)  # 1100 = 12
print('Right shift:', x >> 1)  # 011  = 3
````

````python
# Assignment operators
n = 10
print('Start:', n)
n += 5;  print('+= 5 :', n)
n -= 3;  print('-= 3 :', n)
n *= 2;  print('*= 2 :', n)
n //= 4; print('//= 4:', n)
n **= 2; print('**= 2:', n)
n %= 7;  print('%= 7 :', n)

# Identity and membership operators
my_list = [1, 2, 3, 4, 5]
x = my_list
y = [1, 2, 3, 4, 5]

print('x is my_list:', x is my_list)    # True -- same object
print('y is my_list:', y is my_list)    # False -- different object
print('3 in my_list:', 3 in my_list)    # True
print('9 not in my_list:', 9 not in my_list)  # True
````

## 📘 Day 6 — Tuples

Tuples are like lists but immutable (can't be changed after creation). I use them for things that shouldn't change - like coordinates or RGB colours.

````python
# Creating tuples
empty_tuple = ()
single = (19,)           # note the comma! without it, it's just parentheses
my_info = ('Haney Patel', 19, 'India', 'CS')
coordinates = (19.0760, 72.8777)  # Mumbai lat/lon

print(empty_tuple)
print(single, type(single))
print(my_info)
print('Mumbai coordinates:', coordinates)
````

````python
# Accessing tuple elements
my_info = ('Haney Patel', 19, 'Kenya', 'Data Science')

print('Name:', my_info[0])
print('Age:', my_info[1])
print('Country:', my_info[2])
print('Major:', my_info[3])
print('Last item:', my_info[-1])
print('Slice:', my_info[1:3])

# Tuple unpacking -- very useful!
name, age, country, major = my_info
print(f'{name} is {age} from {country}, studying {major}')

# Swap using tuple packing/unpacking
a, b = 5, 10
a, b = b, a
print(f'a={a}, b={b}')
````

````python
# Tuple methods (only 2!)
numbers = (1, 2, 3, 2, 4, 2, 5)

print('Count of 2:', numbers.count(2))
print('Index of 3:', numbers.index(3))

# Tuples are immutable -- this would raise a TypeError:
# numbers[0] = 99

# But you can reassign the variable
numbers = numbers + (6, 7)
print('Extended tuple:', numbers)

# Useful tuple applications
# 1. Returning multiple values from a function
def min_max(lst):
    return min(lst), max(lst)  # returns a tuple

scores = [78, 92, 65, 88, 71, 95, 83]
lo, hi = min_max(scores)
print(min_max(scores))
print(f'Min: {lo}, Max: {hi}')

# 2. Named tuple -- like a lightweight class
from collections import namedtuple

Student = namedtuple('Student', ['name', 'age', 'major', 'gpa'])
haney = Student('Haney Patel', 19, 'CS', 3.75)
print(haney)
print('Name:', haney.name)
print('GPA:', haney.gpa)
````

````python
# Tuples vs Lists -- when to use which?

# Use a LIST when:
# - Items may change (add/remove/modify)
# - Order might change (sorting)
shopping_cart = ['milk', 'bread', 'eggs']
shopping_cart.append('butter')

# Use a TUPLE when:
# - Data is fixed / shouldn't change
# - Using as dict keys (lists can't be dict keys!)
# - Returning multiple values from a function

# Tuples as dict keys (lists can't do this!)
locations = {
    (19.0760, 72.8777): 'Mumbai',
    (28.6139, 77.2090): 'New Delhi',
    (12.9716, 77.5946): 'Bangalore',
}
print('Location lookup:', locations[(19.0760, 72.8777)])

# Iterating over a tuple
rgb_red   = (255, 0, 0)
rgb_green = (0, 255, 0)
rgb_blue  = (0, 0, 255)

for colour, rgb in [('Red', rgb_red), ('Green', rgb_green), ('Blue', rgb_blue)]:
    print(f'{colour}: R={rgb[0]}, G={rgb[1]}, B={rgb[2]}')
````

````python
# Nested conditionals based on comparisons 
score = 78

if score >= 90:
    grade = 'A'
elif score >= 80:
    grade = 'B'
elif score >= 70:
    grade = 'C'
elif score >= 60:
    grade = 'D'
else:
    grade = 'F'

print(f'Score: {score}, Grade: {grade}')


def max_num(a,b,c):
    if a>=b and b>=c:
        return a
    elif b>=a and b>=c:
        return b
    else:
        return c
    
print(max_num(10,30,50))
````

````python
# Practice: Letter grade calculator function
def get_grade(score):
    if score < 0 or score > 100:
        return 'Invalid score'
    elif score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

test_scores = [95, 83, 72, 61, 55, 101, -5]
for s in test_scores:
    print(f'Score {s}: {get_grade(s)}')
````

````python
# Closures and decorators -- more advanced stuff
def outer(x):
    def inner(y):
        return x + y
    return inner

add5 = outer(5)
print('add5(3):', add5(3))
print('add5(10):', add5(10))

# Simple decorator
def my_decorator(func):
    def wrapper(*args, **kwargs):
        print(f'[LOG] Calling {func.__name__}')
        result = func(*args, **kwargs)
        print(f'[LOG] {func.__name__} returned {result}')
        return result
    return wrapper

@my_decorator
def add(a, b):
    return a + b

add(3, 4)
````

````python
# Custom exceptions
class AgeError(Exception):
    '''Raised when age is invalid'''
    pass

class StudentError(Exception):
    '''Raised for student-related errors'''
    def __init__(self, message, student_name):
        self.student_name = student_name
        super().__init__(f'[{student_name}] {message}')

def validate_age(age):
    if not isinstance(age, int):
        raise TypeError('Age must be an integer')
    if age < 0 or age > 150:
        raise AgeError(f'Invalid age: {age}')
    return True

try:
    validate_age(19)
    print('Age 19 is valid')
    validate_age(-5)
except AgeError as e:
    print('AgeError:', e)
except TypeError as e:
    print('TypeError:', e)

try:
    raise StudentError('Missing assignment submission', 'Haney Patel')
except StudentError as e:
    print('StudentError:', e)
````

````python
# Basic class
class Student:
    school = 'Python University'  # class attribute

    def __init__(self, name, age, major):
        self.name = name
        self.age = age
        self.major = major
        self.grades = {}

    def add_grade(self, subject, grade):
        self.grades[subject] = grade

    def get_gpa(self):
        if not self.grades:
            return 0
        grade_map = {'A+':4.0,'A':4.0,'A-':3.7,'B+':3.3,'B':3.0,'B-':2.7,'C+':2.3,'C':2.0,'D':1.0,'F':0.0}
        points = [grade_map.get(g, 0) for g in self.grades.values()]
        return sum(points) / len(points)

    def __str__(self):
        return f'Student({self.name}, age={self.age}, major={self.major})'

    def __repr__(self):
        return self.__str__()


# Creating my own student object
haney = Student('Haney Patel', 19, 'Computer Science')
haney.add_grade('Python', 'A')
haney.add_grade('Calculus', 'B+')
haney.add_grade('Data Structures', 'A-')

print(haney)
print('Grades:', haney.grades)
print(f'GPA: {haney.get_gpa():.2f}')
print('School:', Student.school)
````

````python
# Inheritance
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        return f'Hi, I am {self.name}, {self.age} years old.'

    def __str__(self):
        return f'Person({self.name}, {self.age})'


class Student(Person):
    def __init__(self, name, age, student_id, major):
        super().__init__(name, age)
        self.student_id = student_id
        self.major = major

    def introduce(self):
        base = super().introduce()
        return f'{base} I study {self.major} (ID: {self.student_id}).'

    def __str__(self):
        return f'Student({self.name}, {self.age}, {self.major})'


class GraduateStudent(Student):
    def __init__(self, name, age, student_id, major, thesis_topic):
        super().__init__(name, age, student_id, major)
        self.thesis_topic = thesis_topic

    def introduce(self):
        base = super().introduce()
        return f'{base} My thesis is on: {self.thesis_topic}.'


haney = Student('Haney Patel', 19, 'CS2026HP', 'Computer Science')
print(haney.introduce())

senior = GraduateStudent('Priya Sharma', 24, 'MS2024PS', 'AI', 'Deep Learning for Medical Imaging')
print(senior.introduce())
````

````python
# Magic methods (dunder methods)
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f'Vector({self.x}, {self.y})'

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def __mul__(self, scalar):
        return Vector(self.x * scalar, self.y * scalar)

    def __len__(self):
        import math
        return int(math.sqrt(self.x**2 + self.y**2))

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y


v1 = Vector(3, 4)
v2 = Vector(1, 2)
print('v1:', v1)
print('v2:', v2)
print('v1 + v2:', v1 + v2)
print('v1 - v2:', v1 - v2)
print('v1 * 3:', v1 * 3)
print('len(v1):', len(v1))
````

````python
# Parsing HTML with a simple example using html.parser (built-in)
from html.parser import HTMLParser

class SimpleHTMLParser(HTMLParser):
    def __init__(self):
        super().__init__()
        self.links = []
        self.headings = []

    def handle_starttag(self, tag, attrs):
        if tag == 'a':
            for attr in attrs:
                if attr[0] == 'href':
                    self.links.append(attr[1])
        if tag in ('h1', 'h2', 'h3'):
            self.current_tag = tag

    def handle_data(self, data):
        if hasattr(self, 'current_tag') and data.strip():
            self.headings.append(data.strip())
            del self.current_tag

sample_html = '''
<html>
  <h1>Haney Patel's Python Notes</h1>
  <h2>Day 22 - Web Scraping</h2>
  <p>Learning web scraping with Python.</p>
  <a href="https://python.org">Python Official Site</a>
  <a href="https://github.com">GitHub</a>
</html>
'''

parser = SimpleHTMLParser()
parser.feed(sample_html)
print('Headings found:', parser.headings)
print('Links found:', parser.links)
````

````python
import statistics

# Sample data -- exam scores from my study group
scores = [72, 85, 90, 88, 76, 92, 68, 84, 79, 95, 88, 71, 83, 91, 77]

print('Count:', len(scores))
print('Sum:', sum(scores))
print('Min:', min(scores))
print('Max:', max(scores))
print('Range:', max(scores) - min(scores))

print('\n--- Central Tendency ---')
print('Mean:', statistics.mean(scores))
print('Median:', statistics.median(scores))
try:
    print('Mode:', statistics.mode(scores))
except statistics.StatisticsError:
    print('Mode: No unique mode')

print('\n--- Spread ---')
print('Variance:', round(statistics.variance(scores), 2))
print('Std Dev:', round(statistics.stdev(scores), 2))
````

````python
# Frequency distribution
from collections import Counter

scores = [72, 85, 90, 88, 76, 92, 68, 84, 79, 95, 88, 71, 83, 91, 77]

def grade_bucket(score):
    if score >= 90: return 'A (90-100)'
    elif score >= 80: return 'B (80-89)'
    elif score >= 70: return 'C (70-79)'
    elif score >= 60: return 'D (60-69)'
    else: return 'F (<60)'

buckets = [grade_bucket(s) for s in scores]
freq = Counter(buckets)

print('Grade Distribution:')
for grade in ['A (90-100)', 'B (80-89)', 'C (70-79)', 'D (60-69)', 'F (<60)']:
    count = freq.get(grade, 0)
    bar = '#' * count
    print(f'{grade}: {bar} ({count})')
````

````python
# Percentiles and quartiles
import statistics

scores = sorted([72, 85, 90, 88, 76, 92, 68, 84, 79, 95, 88, 71, 83, 91, 77])

n = len(scores)
q1 = statistics.median(scores[:n//2])
q3 = statistics.median(scores[(n+1)//2:])
iqr = q3 - q1

print('Sorted scores:', scores)
print(f'Q1: {q1}')
print(f'Q2 (Median): {statistics.median(scores)}')
print(f'Q3: {q3}')
print(f'IQR: {iqr}')

# Find outliers (below Q1-1.5*IQR or above Q3+1.5*IQR)
lower_fence = q1 - 1.5 * iqr
upper_fence = q3 + 1.5 * iqr
outliers = [s for s in scores if s < lower_fence or s > upper_fence]
print(f'Outlier fences: [{lower_fence:.1f}, {upper_fence:.1f}]')
print('Outliers:', outliers if outliers else 'None')
````

````python
# Complete API client class
import json

class APIClient:
    '''Simple API client for practicing REST concepts'''

    def __init__(self, base_url):
        self.base_url = base_url
        self._mock_data = {
            'users': [
                {'id': 1, 'name': 'Haney Patel', 'age': 19},
                {'id': 2, 'name': 'Priya Singh', 'age': 20},
            ]
        }

    def get(self, endpoint):
        # In real use: response = requests.get(f'{self.base_url}/{endpoint}')
        resource = endpoint.strip('/')
        return self._mock_data.get(resource, [])

    def post(self, endpoint, data):
        resource = endpoint.strip('/')
        if resource in self._mock_data:
            new_item = {'id': len(self._mock_data[resource]) + 1, **data}
            self._mock_data[resource].append(new_item)
            return new_item
        return None

    def delete(self, endpoint, item_id):
        resource = endpoint.strip('/')
        if resource in self._mock_data:
            original = len(self._mock_data[resource])
            self._mock_data[resource] = [x for x in self._mock_data[resource] if x['id'] != item_id]
            return len(self._mock_data[resource]) < original
        return False


client = APIClient('https://api.example.com')

print('GET /users:', client.get('/users'))
new_user = client.post('/users', {'name': 'Arjun Mehta', 'age': 18})
print('POST new user:', new_user)
print('After POST:', client.get('/users'))
deleted = client.delete('/users', 1)
print('DELETE user 1:', deleted)
print('After DELETE:', client.get('/users'))
````

## 📘 Day 29 — Building a REST API with Flask

Building a full CRUD API. This is the project I'm most proud of from this challenge!


````python
# Full REST API code (save as app.py and run with python app.py)

full_api_code = '''
from flask import Flask, jsonify, request
from datetime import datetime
import uuid

app = Flask(__name__)

# In-memory store
db = {
    'students': [
        {
            'id': str(uuid.uuid4()),
            'name': 'Haney Patel',
            'age': 19,
            'major': 'Computer Science',
            'enrolled': True,
            'created_at': datetime.now().isoformat()
        }
    ]
}

def not_found(msg):
    return jsonify({'error': msg, 'status': 404}), 404

def bad_request(msg):
    return jsonify({'error': msg, 'status': 400}), 400

# ---- STUDENTS ENDPOINTS ----

@app.route('/api/v1/students', methods=['GET'])
def get_students():
    major = request.args.get('major')
    students = db['students']
    if major:
        students = [s for s in students if s['major'].lower() == major.lower()]
    return jsonify({'data': students, 'count': len(students), 'status': 200})

@app.route('/api/v1/students/<student_id>', methods=['GET'])
def get_student(student_id):
    student = next((s for s in db['students'] if s['id'] == student_id), None)
    if not student:
        return not_found(f'Student {student_id} not found')
    return jsonify({'data': student, 'status': 200})

@app.route('/api/v1/students', methods=['POST'])
def create_student():
    data = request.get_json()
    required = ['name', 'age', 'major']
    missing = [f for f in required if f not in data]
    if missing:
        return bad_request(f'Missing required fields: {missing}')

    student = {
        'id': str(uuid.uuid4()),
        'name': data['name'],
        'age': data['age'],
        'major': data['major'],
        'enrolled': data.get('enrolled', True),
        'created_at': datetime.now().isoformat()
    }
    db['students'].append(student)
    return jsonify({'data': student, 'status': 201}), 201

@app.route('/api/v1/students/<student_id>', methods=['PUT'])
def update_student(student_id):
    student = next((s for s in db['students'] if s['id'] == student_id), None)
    if not student:
        return not_found(f'Student {student_id} not found')
    data = request.get_json()
    for key in ['name', 'age', 'major', 'enrolled']:
        if key in data:
            student[key] = data[key]
    student['updated_at'] = datetime.now().isoformat()
    return jsonify({'data': student, 'status': 200})

@app.route('/api/v1/students/<student_id>', methods=['DELETE'])
def delete_student(student_id):
    original_len = len(db['students'])
    db['students'] = [s for s in db['students'] if s['id'] != student_id]
    if len(db['students']) == original_len:
        return not_found(f'Student {student_id} not found')
    return jsonify({'message': 'Deleted successfully', 'status': 200})

if __name__ == '__main__':
    app.run(debug=True, port=5000)
'''

print('Full REST API code:')
print(full_api_code)
````
